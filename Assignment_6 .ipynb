9. What are the methods for right-justifying, left-justifying, and centering a string?
Ans: rjust(),ljust(),center()

```python
"I Love India".rjust(20,"*")
```

```python
"I Love India".ljust(20,"*")
```

```python
"I Love India".center(20,"*")
```
